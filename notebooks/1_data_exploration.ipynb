```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')
```

```python
df = pd.read_csv('../data/raw_data/creditcard.csv')
df.head()   
```

```python
print('Shape: ', df.shape)
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
class_counts = df['Class'].value_counts()

print(class_counts)
print("\nClass Distribution (%): ")
print(class_counts / len(df) * 100)
```

```python
sns.countplot(x='Class', data=df)
plt.title('Class Distribution (Fraud vs Non-Fraud)')
plt.xlabel('Class (0: Non-Fraud, 1: Fraud)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df[df['Class'] == 0]['Amount'], bins=50, color='green', label='Non-Fraud', stat='density')
sns.histplot(df[df['Class'] == 1]['Amount'], bins=50, color='red', label='Fraud', stat='density')
plt.legend()
plt.title('Transaction Amount Distribution by Class')
plt.xlabel('Transaction Amount')
plt.ylabel('Density')
plt.show()  
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['Time'], bins=50, color='blue', stat='density', kde=True)
plt.title('Transaction Time Distribution')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), cbar=False, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
